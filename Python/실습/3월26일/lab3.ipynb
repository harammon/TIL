```python
import requests
from bs4 import BeautifulSoup
import re

url = 'http://media.daum.net/ranking/popular/'
r = requests.get(url)
r.encoding = 'utf-8'
bs = BeautifulSoup(r.text, 'html.parser')
newstitle=[]
newscomname=[]
newstitleList = bs.select('#mArticle > div.rank_news > ul.list_news2 > li > div.cont_thumb > strong > a')
newscomnameList = bs.select('#mArticle > div.rank_news > ul.list_news2 > li > div.cont_thumb > strong > span')
for newstitleDom in newstitleList:
    newstitleDom=re.sub(", ", ' ', newstitleDom.text)
    newstitle.append(newstitleDom)
    print(newstitle)
for newscomnameDom in newscomnameList:
    newscomname.append(newscomnameDom.string)
    print(newscomnameDom.string)
#print(title)
with open('output/news.csv', "wt", encoding="utf-8") as f: # wt : 텍스트 기반의 출력 모드로 open!
    f.write('newstitle,newscomname\n') # 글의 맨 위에 작성될 것임  
    for i in range(len(newstitle)):
        f.write(newstitle[i]+","+newscomname[i]+'\n')  
```
